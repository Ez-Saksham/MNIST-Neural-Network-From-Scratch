```python
def init_Para():
    #for 1st layer
    W1 = np.random.rand(10, 784) #10 rows and 784 col having 784 weights for each rows (neurons) i.e 10
    b1 = np.random.rand(10, 1) #10 rows with 1 col having 1 bias for each rows (neurons)
    #for 2nd layer 
    W2 = np.random.rand(10, 10)
    b2 = np.random.rand(10, 1)
    return W1,b1,W2,b2
```

```python
def sigmoid(Zi): #activation function
    return 1 / (1 + np.exp(-Zi))
```

```python
def softmax(Zi):
    return exp(Zi)/np.sum(exp(Zi))
```

```python
def forward_pass(W1,b1,W2,b2,X):
    Z1 = W1.dot(X) + b1
    A1 = sigmoid(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = softmax(A1)
    return Z1,A1,Z2,A2
```
